# **PROJECT-TITANIC SURVIVORS ANALYSIS**

```python
import pandas as pd
import numpy as np
```

## READING DATA IN TO PANDAS

Use ***read_csv*** function to read the csv file in python and pandas ***DataFrame*** method to convert file in to data frame.

```python
df = pd.DataFrame(pd.read_csv('/content/train (1).csv'))
df.head()
```

```python
df.describe()
```

```python
df.shape
```

## HANDELING NULL VALUES

```python
#finging the null values
y = df.isnull().sum()
y
```

```python
#Just drop the column which have more than 35% of the data missing.
drop_col = y[y>(35/100 * df.shape[0])]
drop_col
```

```python
df.shape
```

```python
drop_col.index
```

```python
df.drop(drop_col.index, axis=1, inplace=True)
df.isnull().sum()
```

```python
# column EMBARKED contains string values
df['Embarked'].fillna('S',inplace=True)
```

```python
df.isnull().sum()
```

```python
#handeling null values replaced by the mean of the column.
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

## CORRELATION OF THE ALL COLUMNS OF THE DATAFRAME

                                                                                                       

 To find the correlation of all columns in the dataframe we use **df.corr() **function

```python
df.corr()
```

```python
#group 'SibSp'and 'Parch' in to new column 'famsize'
df["famsize"] = df["SibSp"]+df["Parch"]
df.drop(['SibSp'], axis=1, inplace=True)
df.drop(['Parch'], axis=1, inplace=True)
df.corr()
```

### Finding survivors with no family

```python
#creating a column 'NoFam' of people who has no family
df['NoFam']=[0 if df['famsize'][i]>0 else 1 for i in df.index]
df.head()
```

```python
#Checking the chance of surviving of person who has no family
df.groupby(['NoFam'])['Survived'].mean()
```

This means if the person is alone who has less chance of surviving!

```python
df[['NoFam','Fare']].corr()
```

so the ticket price for lonely person is high.

```python
df['sex']=[0 if df['Sex'][i]=='male' else 1 for i in df.index]
df.groupby(['sex'])['Survived'].mean()
```

so female passangers have more chance of surviving than males.
That is women where prioritized.

```python
#checking th survival chances of embarked passangers.
df.groupby(['Embarked'])['Survived'].mean()
```

# **CONCLUSION**


1.   Passangers who are **alone** have **less chance** of survival rate.
2.   The passangers who are **rich** **and have family** have **more** chance to get survived **than** the **poor** alone passangers.
3.   **Female** passangers get prioritized over **male **passangers.
4.   The survivors who **borded the ship** have more survival rate than others.
